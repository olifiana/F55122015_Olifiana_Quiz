```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageOps
from skimage.util import random_noise
```

```python
# Load image using OpenCV
image_path = ('/content/DN Motor.jpeg')
image_cv = cv2.imread(image_path)
```

```python
# Display original image
plt.imshow(cv2.cvtColor(image_cv, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
# Edge detection using Sobel
def sobel_edge_detection(image):
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    sobelx = cv2.Sobel(gray_image, cv2.CV_64F, 1, 0, ksize=5)
    sobely = cv2.Sobel(gray_image, cv2.CV_64F, 0, 1, ksize=5)
    sobel_combined = cv2.magnitude(sobelx, sobely)
    return sobel_combined
    sobel_image = sobel_edge_detection(image_cv)
```

```python
# Display Sobel edge detection result
plt.imshow(sobel_image, cmap='gray')
plt.title('Sobel Edge Detection')
plt.axis('off')
plt.show()
```

```python
# Edge detection using Canny
def canny_edge_detection(image):
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    canny_edges = cv2.Canny(gray_image, 100, 200)
    return canny_edges

canny_image = canny_edge_detection(image_cv)
```

```python
# Display Canny edge detection result
plt.imshow(canny_image, cmap='gray')
plt.title('Canny Edge Detection')
plt.axis('off')
plt.show()
```

```python
# Random cropping
def random_crop(image):
    h, w = image.shape[:2]
    x = np.random.randint(0, w // 2)
    y = np.random.randint(0, h // 2)
    cropped_image = image[y:y + h // 2, x:x + w // 2]
    return cropped_image

random_cropped_image = random_crop(image_cv)
```

```python
# Display random cropped image
plt.imshow(cv2.cvtColor(random_cropped_image, cv2.COLOR_BGR2RGB))
plt.title('Random Cropped Image')
plt.axis('off')
plt.show()
```

```python
# Half-size center cropping
def center_crop(image):
    h, w = image.shape[:2]
    center_x, center_y = w // 2, h // 2
    cropped_image = image[center_y - h // 4:center_y + h // 4, center_x - w // 4:center_x + w // 4]
    return cropped_image

center_cropped_image = center_crop(image_cv)
```

```python
# Display half-size center cropped image
plt.imshow(cv2.cvtColor(center_cropped_image, cv2.COLOR_BGR2RGB))
plt.title('Half-size Center Cropped Image')
plt.axis('off')
plt.show()
```

```python
# Invert colors (white to black, black to white)
inverted_image = cv2.bitwise_not(image_cv)
```

```python
# Display inverted image
plt.imshow(cv2.cvtColor(inverted_image, cv2.COLOR_BGR2RGB))
plt.title('Inverted Image')
plt.axis('off')
plt.show()
```

```python
# Add salt and pepper noise
noisy_image_salt_pepper = random_noise(image_cv, mode='s&p', amount=0.05)
```

```python
# Display image with salt and pepper noise
plt.imshow(cv2.cvtColor((noisy_image_salt_pepper * 255).astype(np.uint8), cv2.COLOR_BGR2RGB))
plt.title('Image with Salt and Pepper Noise')
plt.axis('off')
plt.show()
```

```python
# Add Gaussian noise
noisy_image_gaussian = random_noise(image_cv, mode='gaussian', var=0.01**2)
```

```python
# Display image with Gaussian noise
plt.imshow(cv2.cvtColor((noisy_image_gaussian * 255).astype(np.uint8), cv2.COLOR_BGR2RGB))
plt.title('Image with Gaussian Noise')
plt.axis('off')
plt.show()
```

```python
# Geometric transformations: rotation, vertical flip, horizontal flip
# Rotation
rotation_angle = 30
rows, cols = image_cv.shape[:2]
rotation_matrix = cv2.getRotationMatrix2D((cols/2, rows/2), rotation_angle, 1)
rotated_image = cv2.warpAffine(image_cv, rotation_matrix, (cols, rows))
```

```python
# Display rotated image
plt.imshow(cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB))
plt.title('Rotated Image')
plt.axis('off')
plt.show()
```

```python
# Vertical flip
vertical_flipped_image = cv2.flip(image_cv, 0)
```

```python
# Display vertically flipped image
plt.imshow(cv2.cvtColor(vertical_flipped_image, cv2.COLOR_BGR2RGB))
plt.title('Vertically Flipped Image')
plt.axis('off')
plt.show()
```

```python
# Horizontal flip
horizontal_flipped_image = cv2.flip(image_cv, 1)
```

```python
# Display horizontally flipped image
plt.imshow(cv2.cvtColor(horizontal_flipped_image, cv2.COLOR_BGR2RGB))
plt.title('Horizontally Flipped Image')
plt.axis('off')
plt.show()
```
